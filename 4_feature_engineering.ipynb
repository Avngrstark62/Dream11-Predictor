```python
import pandas as pd
```

# read data

```python
data = pd.read_csv('data/final_preprocessed_dataset.csv')
data['match_date'] = pd.to_datetime(data['match_date'], format="%d-%m-%Y")
data
```

```python
data.info()
```

# opponent_team_name

```python
def get_opponent_team_name(row):
    team_name = row['team_name']
    team1_name = row['team1_name']
    team2_name = row['team2_name']
    if team_name == team1_name:
        return team2_name
    elif team_name == team2_name:
        return team1_name
    else:
        return None

data['opponent_team_name'] = data.apply(get_opponent_team_name, axis=1)
data
```

# team_score and opponent_team_score

```python
def get_team_score(row):
    team_name = row['team_name']
    team1_name = row['team1_name']
    team2_name = row['team2_name']
    team1_score = row['team1_score']
    team2_score = row['team2_score']
    if team_name == team1_name:
        return team1_score
    elif team_name == team2_name:
        return team2_score
    else:
        return None

def get_opponent_team_score(row):
    opponent_team_name = row['opponent_team_name']
    team1_name = row['team1_name']
    team2_name = row['team2_name']
    team1_score = row['team1_score']
    team2_score = row['team2_score']
    if opponent_team_name == team1_name:
        return team1_score
    elif opponent_team_name == team2_name:
        return team2_score
    else:
        return None

data['team_score'] = data.apply(get_team_score, axis=1)
data['opponent_team_score'] = data.apply(get_opponent_team_score, axis=1)
data
```

# drop team1_name, team2_name, team1_score and team2_score

```python
data = data.drop(columns=['team1_name', 'team2_name', 'team1_score', 'team2_score'])
data
```

# team_won

```python
def team_won(row):
    match_winner = row['match_winner']
    team_name = row['team_name']
    opponent_team_name = row['opponent_team_name']
    if match_winner == team_name:
        return 2
    elif match_winner == opponent_team_name:
        return 0
    elif match_winner in ['Match tied', 'No result']:
        return 1
    else:
        return None

data['team_won'] = data.apply(team_won, axis=1)
data = data.drop(columns=['match_winner'])
data
```

# batting_first

```python
def batting_first(row):
    team_name = row['team_name']
    opponent_team_name = row['opponent_team_name']
    toss_winner = row['toss_winner']
    toss_decision = row['toss_decision']

    if toss_winner == team_name:
        if toss_decision == 'bat':
            return 1
        elif toss_decision == 'field':
            return 0
        else:
            return None
    elif toss_winner == opponent_team_name:
        if toss_decision == 'bat':
            return 0
        elif toss_decision == 'field':
            return 1
        else:
            return None
    else:
        return None

data['batting_first'] = data.apply(batting_first, axis=1)
data = data.drop(columns=['toss_winner', 'toss_decision'])
data
```

# rearrange_columns

```python
data = data[['performance_id', 'match_id', 'match_date', 'player', 'team_name', 'opponent_team_name', 'ground', 'format',
             'team_score', 'opponent_team_score', 'batting_first', 'team_won', 'runs', 'wickets']]
data
```

# recent_team_scores_avg

```python
def get_recent_team_scores_avg(data):
    df = data[['match_id', 'match_date', 'team_name', 'team_score']].sort_values(by=['match_date', 'team_name'])
    df.drop_duplicates(subset=None, keep='first', inplace=True)
    df['recent_team_scores_avg'] = df.groupby('team_name')['team_score'].transform(lambda x: x.shift().rolling(window=3, min_periods=1).mean())
    new_data = pd.merge(data, df[['match_id', 'team_name', 'recent_team_scores_avg']], on=['match_id', 'team_name'], how='outer')
    return new_data

data = get_recent_team_scores_avg(data)
data
```

# recent_team_win_rate

```python
def get_recent_team_win_rate(data):
    df = data[['match_id', 'match_date', 'team_name', 'team_won']].sort_values(by=['match_date', 'team_name'])
    df.drop_duplicates(subset=None, keep='first', inplace=True)
    df['recent_team_win_rate'] = df.groupby('team_name')['team_won'].transform(lambda x: x.shift().rolling(window=3, min_periods=1).mean())
    new_data = pd.merge(data, df[['match_id', 'team_name', 'recent_team_win_rate']], on=['match_id', 'team_name'], how='outer')
    return new_data

data = get_recent_team_win_rate(data)
data
```

# recent_runs_avg

```python
def get_recent_runs_avg(data):
    df = data[['match_id', 'match_date', 'player', 'team_name', 'runs']].sort_values(by=['match_date', 'team_name'])
    df['recent_runs_avg'] = df.groupby('player')['runs'].transform(lambda x: x.shift().rolling(window=3, min_periods=1).mean())
    new_data = pd.merge(data, df[['match_id', 'team_name', 'player', 'recent_runs_avg']], on=['match_id', 'team_name', 'player'], how='outer')
    return new_data

data = get_recent_runs_avg(data)
data
```

# recent_wickets_avg

```python
def get_recent_wickets_avg(data):
    df = data[['match_id', 'match_date', 'player', 'team_name', 'wickets']].sort_values(by=['match_date', 'team_name'])
    df['recent_wickets_avg'] = df.groupby('player')['wickets'].transform(lambda x: x.shift().rolling(window=3, min_periods=1).mean())
    new_data = pd.merge(data, df[['match_id', 'team_name', 'player', 'recent_wickets_avg']], on=['match_id', 'team_name', 'player'], how='outer')
    return new_data

data = get_recent_wickets_avg(data)
data
```

# fantasy_score

```python
data['fantasy_score'] = data['runs'] + 25 * data['wickets']
data
```

# drop unneccesary columns

```python
data = data.drop(columns=['team_score', 'opponent_team_score', 'team_won', 'runs', 'wickets'])
data
```

# labelling

```python
from sklearn.preprocessing import LabelEncoder

le_player = LabelEncoder()
data['player_label'] = le_player.fit_transform(data['player'])

le_team = LabelEncoder()
data['team_name_label'] = le_team.fit_transform(data['team_name'])
data['opponent_team_name_label'] = le_team.transform(data['opponent_team_name'])

le_ground = LabelEncoder()
data['ground_label'] = le_ground.fit_transform(data['ground'])


data = data.drop(columns=['team_name', 'ground', 'opponent_team_name'])
data
```

# format

```python
def format_t20(format_):
    if format_ == 20:
        return 1
    else:
        return 0

def format_odi(format_):
    if format_ == 50:
        return 1
    else:
        return 0

data['format_t20'] = data['format'].apply(format_t20)
data['format_odi'] = data['format'].apply(format_odi)
data = data.drop(columns=['format'])
data
```

# handle null values

```python
data = data.dropna()
data
```

# top_performer

```python
def top_performer(row, data):
    match_id = row['match_id']
    match_data = data[data['match_id'] == match_id]
    rank_list = match_data.sort_values(by=['fantasy_score'])['player'].tolist()
    rank_list.reverse()
    
    player = row['player']
    player_rank = (rank_list.index(player)) + 1

    if player_rank==1:
        return 4
    elif player_rank==2:
        return 2
    elif player_rank<=11:
        return 1
    else:
        return 0

data['top_performer'] = data.apply(lambda x: top_performer(x, data), axis=1)
data
```
