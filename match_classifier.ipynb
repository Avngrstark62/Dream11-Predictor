# opponent_team_name

```python
def get_opponent_team_name(row):
    team_name = row['team_name']
    team1_name = row['team1_name']
    team2_name = row['team2_name']
    if team_name == team1_name:
        return team2_name
    elif team_name == team2_name:
        return team1_name
    else:
        return None

data['opponent_team_name'] = data.apply(get_opponent_team_name, axis=1)
data
```

# team_score and opponent_team_score

```python
def get_team_score(row):
    team_name = row['team_name']
    team1_name = row['team1_name']
    team2_name = row['team2_name']
    team1_score = row['team1_score']
    team2_score = row['team2_score']
    if team_name == team1_name:
        return team1_score
    elif team_name == team2_name:
        return team2_score
    else:
        return None

def get_opponent_team_score(row):
    opponent_team_name = row['opponent_team_name']
    team1_name = row['team1_name']
    team2_name = row['team2_name']
    team1_score = row['team1_score']
    team2_score = row['team2_score']
    if opponent_team_name == team1_name:
        return team1_score
    elif opponent_team_name == team2_name:
        return team2_score
    else:
        return None

data['team_score'] = data.apply(get_team_score, axis=1)
data['opponent_team_score'] = data.apply(get_opponent_team_score, axis=1)
data
```

# drop team1_name, team2_name, team1_score and team2_score

```python
data = data.drop(columns=['team1_name', 'team2_name', 'team1_score', 'team2_score'])
data
```

# team_won

```python
def team_won(row):
    match_winner = row['match_winner']
    team_name = row['team_name']
    opponent_team_name = row['opponent_team_name']
    if match_winner == team_name:
        return 2
    elif match_winner == opponent_team_name:
        return 0
    elif match_winner in ['Match tied', 'No result']:
        return 1
    else:
        return None

data['team_won'] = data.apply(team_won, axis=1)
data = data.drop(columns=['match_winner'])
data
```

# batting_first

```python
def batting_first(row):
    team_name = row['team_name']
    opponent_team_name = row['opponent_team_name']
    toss_winner = row['toss_winner']
    toss_decision = row['toss_decision']

    if toss_winner == team_name:
        if toss_decision == 'bat':
            return 1
        elif toss_decision == 'field':
            return 0
        else:
            return None
    elif toss_winner == opponent_team_name:
        if toss_decision == 'bat':
            return 0
        elif toss_decision == 'field':
            return 1
        else:
            return None
    else:
        return None

data['batting_first'] = data.apply(batting_first, axis=1)
data = data.drop(columns=['toss_winner', 'toss_decision'])
data
```

# match_classifier

```python
def get_match_classifier_data(match_id, data, match_classifier_data):
    match_classifier_df = match_classifier_data.copy()
    
    match_data = data[data['match_id']==match_id]
    ground = match_data['ground'].unique()[0]
    match_format = match_data['format'].unique()[0]
    
    team1_data = match_data[match_data['batting_first']==1].sort_values(by='performance_id')[['team_name', 'team_score', 'runs']]
    team1_name = team1_data['team_name'].unique()[0]
    team1_score = team1_data['team_score'].unique()[0]
    team1_runs = team1_data['runs'].tolist()
    
    team2_data = match_data[match_data['batting_first']==0].sort_values(by='performance_id')[['team_name', 'team_score', 'runs']]
    team2_name = team2_data['team_name'].unique()[0]
    team2_score = team2_data['team_score'].unique()[0]
    team2_runs = team2_data['runs'].tolist()
    
    match_dict = {
        'match_id': [match_id],
        'ground': [ground],
        'format': [match_format],
        'team1_name': [team1_name],
        'team1_score': [team1_score],
        'team2_name': [team2_name],
        'team2_score': [team2_score],
    }
    for i in range(11):
        try:
            match_dict[f'team{1}_player{i+1}'] = [team1_runs[i]]
        except IndexError:
            match_dict[f'team{1}_player{i+1}'] = None
    for i in range(11):
        try:
            match_dict[f'team{2}_player{i+1}'] = [team2_runs[i]]
        except IndexError:
            match_dict[f'team{2}_player{i+1}'] = None
    # return match_dict
    
    match_df = pd.DataFrame(match_dict)
    # return match_df
    return pd.concat([match_classifier_df, match_df], axis=0)
```

```python
cols = ['match_id','ground', 'format', 'team1_name', 'team1_score', 'team2_name', 'team2_score']
for i in range(11):
    cols.append(f'team{1}_player{i+1}')
for i in range(11):
    cols.append(f'team{2}_player{i+1}')
match_classifier_data = pd.DataFrame(columns=cols)
match_classifier_data
```

```python
match_ids = data['match_id'].unique().tolist()
for match_id in match_ids:
    match_classifier_data = get_match_classifier_data(match_id, data, match_classifier_data)
```

# scorecard_similarity

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

def scorecard_similarity_manual(vec1, vec2):
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)
    similarity_score = 0
    for i in range(len(vec1)):
        if abs(vec1[i]-vec2[i])<=10:
            similarity_score+=1
        if vec1[i]==max(vec1) and vec2[i]==max(vec2):
            similarity_score+=5
    return similarity_score

import pandas as pd

def scorecard_similarity_matrix(real_data):
    data = real_data.copy()
    match_ids = data['match_id'].unique()
    data = data.drop(columns=['match_id'])
    
    similarity_matrix = pd.DataFrame(index=match_ids, columns=match_ids)
    
    for i in range(len(match_ids)):
        for j in range(i, len(match_ids)):  # Only compute upper triangle
            sim = scorecard_similarity_manual(data.iloc[i], data.iloc[j])
            similarity_matrix.iloc[i, j] = sim
            similarity_matrix.iloc[j, i] = sim  # Fill symmetric value

    return similarity_matrix
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

# Assume df is your original DataFrame
df = match_classifier_data.copy()

# Step 2: Drop unnecessary columns
data_cleaned = df.drop(columns=['ground', 'format', 'team1_name', 'team2_name', 'team1_score', 'team2_score'])
similarity_matrix = scorecard_similarity_matrix(data_cleaned)
similarity_matrix
```

# analysis

```python
pd.set_option('display.max_columns', None)

def get_similar_matches(match_id, similarity_matrix, top_n=5):
    if match_id not in similarity_matrix.index:
        raise ValueError(f"Match ID {match_id} not found in the similarity matrix.")
    similar_matches = similarity_matrix.loc[match_id].drop(match_id)
    most_similar = similar_matches.sort_values(ascending=False).head(top_n)

    # Return the result as a DataFrame
    return most_similar.reset_index().rename(columns={match_id: 'similarity_value'})

# Get the most similar matches for a specific match_id (e.g., 1)
match_id = 1
similar_matches = get_similar_matches(match_id, similarity_matrix=similarity_matrix, top_n=400)
similar_matches
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

max_value = 27
distance_matrix = max_value - similarity_matrix

# K-Means clustering
k = 10  # Adjust based on your analysis
kmeans = KMeans(n_clusters=k)
kmeans.fit(distance_matrix)
labels = kmeans.labels_

# Results
cluster_results = pd.DataFrame({'match_id': distance_matrix.index, 'cluster': labels})
cluster_results
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

max_value = 32
# Convert similarity to distance
distance_matrix = max_value - similarity_matrix

if (distance_matrix < 0).any().any():
    raise ValueError("The distance matrix contains negative values. Check your similarity matrix.")

threshold = 14
eps = max_value - threshold

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=eps, min_samples=1, metric='precomputed')
labels = dbscan.fit_predict(distance_matrix)

# Results
cluster_results = pd.DataFrame({'match_id': distance_matrix.index, 'cluster': labels})
cluster_results
```
